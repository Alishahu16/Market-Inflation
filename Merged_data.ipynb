```python
import pandas as pd
```

```python
Gas = pd.read_excel('Gas Price.xls')
Egg = pd.read_excel('NewEggPrice.xlsx')
Milk = pd.read_excel('NewMilkPrice.xlsx')
Rice= pd.read_excel('NewRicePrice.xlsx')
```

```python
Gas.head()
```

```python
Gas =Gas.rename(columns={'DATE':'Date'})
```

```python
Gas.head()
```

```python
Egg.head()
```

```python
Milk.head()
```

```python
Rice.head()
```

```python
# Ensure 'Date' columns are in datetime format before merging
Gas['Date'] = pd.to_datetime(Gas['Date'])
Egg['Date'] = pd.to_datetime(Egg['Date'])
Milk['Date'] = pd.to_datetime(Milk['Date'])
Rice['Date'] = pd.to_datetime(Rice['Date'])
```

```python
Gas.head()
```

```python
# Perform the merges sequentially, renaming columns to avoid conflicts
Mergeddata = pd.merge(Gas, Egg, on='Year', how='inner', suffixes=('_gas', '_egg'))
Mergeddata = pd.merge(Mergeddata, Milk, on='Year', how='inner', suffixes=('_egg_milk','_milk'))  # Changed from '_egg' to '_egg_milk'
Mergeddata = pd.merge(Mergeddata, Rice, on='Year', how='inner', suffixes=('_milk_rice', '_rice')) # Changed from '_milk' to '_milk_rice'
```

```python
Mergeddata.head()
```

```python
Mer_data = "Mergedata.xlsx"
Mergeddata.to_excel(Mer_data, index=False)
```
